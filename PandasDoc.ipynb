```python
# Combine the two pandas series 
data1 = [23,45,67,89,23,45,67]
data2 = [ 34, 56,789,23,46,68,22]
s1 = pd.Series(data1)
s2 = pd.Series(data2)
print(s1)
print(s2)
def demo(x1, x2):
  if(x1 > x2):
    return x1
  else:
    return x2
#  combining the two series into one 
# The function return the largest value 
res = s1.combine(s2, demo)
print("After combining the \n" ,res)
```

```python
# Aggregation Operation - Get the mean of the grouped data
data={
    "Player":["Muzamil","Fatima", "Mughal", "Zeeshan ", "Ali" ,"Meerab"],
    "Rank":[1,4,3,5,2,7],
    "points":[95,70,65,80,90,50],
    "Year":[2023,2022,2021,2022,2018,2019]
}
df = pd.DataFrame(data)
print("Cricket Player Records\n", df)
groupRes =df.groupby('Year')
print("\n", groupRes['points'].agg(np.mean))
```

```python
# Get the size of each group (aggregatation)
data={
    "Player":["Muzamil","Fatima", "Mughal", "Zeeshan ", "Ali" ,"Meerab"],
    "Rank":[1,4,3,5,2,7],
    "points":[95,70,65,80,90,50],
    "Year":[2023,2022,2021,2022,2018,2019]
}
df = pd.DataFrame(data)
print("Cricket Player Records\n", df)
groupRes = df.groupby("Player")
print(groupRes.agg(np.size))
```
