```python
# intersect the two numpy array 
n1 = np.array([3,4,5,8,93,839])
n2 = np.array([4,5,6,4,25,245])
print('Iternate first array')
for a in n1:
    print(a)
print('iterate second array')
for a in n2:
    print (a)
resarr = np.intersect1d(n1, n2)
print("intersect between two array is ", resarr)
```
