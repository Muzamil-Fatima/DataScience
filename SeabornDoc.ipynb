```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Line plot in the seaborn
var1 = [1,2,3,4,5,6,7]
var2 = [2,4,5,6,4,8,9]
plt.plot (var1 , var2)
plt.xlabel("X-axis varibale" )
plt.ylabel("Y-axis variable")
plt.show()
```

```python
var_1 = [1,2,3,4,5,6,7]
var_2 = [2,4,5,6,4,8,9]
x_1 =pd.DataFrame({"var_1":var_1 ,"var_2" :var_2})
sns.lineplot(x="var_1", y= "var_2", data=x_1)
plt.show()
```

```python
var_1 = [1,2,3,4,5,6,7]
var_2 = [2,4,5,6,4,8,9]
sns.lineplot(x= var_1, y= var_2)
plt.show()
```
