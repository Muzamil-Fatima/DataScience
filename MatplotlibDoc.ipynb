```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xpts = np.array([0, 10])
ypts = np.array([0, 125])
plt.plot(xpts,ypts)
plt.show()
```

```python
data= {
    'Student': ["Muzamil","Fatima","Mughal","Farhan"],
    'Rank':[1,2,3,4],
}
df = pd.DataFrame(data)
plt.plot(df["Student"], df["Rank"])
plt.title("This is our student data", loc = 'center')
plt.xlabel("Student Name")
plt.ylabel("Rank of Student")
plt.grid()
plt.show()
```

```python
# Create the legend
a = np.arange(5)
b = [2,4,6,8,10]
c = [1,2,3,4,5]
# create a plot
fig = plt.figure()
ax = plt.subplot()
ax.plot(a,b, 'k--', label = 'Frequency')
ax.plot(a,c, 'k:', label = 'periods')
ax.legend()
plt.show()
```

```python
#  after creating the legend we just change the color of legend
a = np.arange(5)
b = [2,4,6,8,10]
c = [1,2,3,4,5]
fig = plt.figure()
ax = plt.subplot()
ax.plot(a,b, 'k--', label='Frequency')
ax.plot(a,c, 'k:', label='Period')
legend = ax.legend(loc = 'upper center')
plt.title("Frequency of signal")
legend.get_frame().set_facecolor('red')
plt.show()
```

```python
student = np.array(["Muzmail", "fatima", "Mughal", "Zubair", "Hussain"])
marks = np.array([86,87,56,87,98,])
plt.bar(student, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Bar graph student Data")
plt.show()
```

```python
cricketer= np.array(["Muzamil", "Fatima", "Mughal", "Mehmoona ", "Asma" , "Farhan"])
runs =np.array([99, 98,95, 89,79,93])
plt.pie(runs, labels = cricketer, autopct = '%1.2f%%')
plt.title("Run Secored in a cricket match")
plt.show()
```

```python
# Draw a histogram
arr= np.array([1,23,44,55,32,34,65,33,22,78,55,43,23,45,67,8,87,23,66])
plt.hist(arr, bins= [0,20,40,60,80])
plt.title("Student Marks")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()
```
